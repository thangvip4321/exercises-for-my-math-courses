```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
def illustration_3d_func(matrix):
    # take a matrix with shape (n_sample,3) as an argument
    #leave the last 2 for different colorization?
    X = matrix[:-2,0]
    Y = matrix[:-2,1]
    Z = matrix[:-2,2]
    X1 =matrix[-2,0]
    Y1 = matrix[-2,1]
    Z1 = matrix[-2,2]
    X2 =matrix[-1,0]
    Y2 = matrix[-1,1]
    Z2 = matrix[-1,2]
    fig = plt.figure(figsize=plt.figaspect(0.4))
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.scatter(X, Y, Z, zdir='z', s=0.1, c='r', depthshade=True)
    ax.scatter(X1, Y1, Z1, zdir='z', s=10, c='g', depthshade=True)
    ax.scatter(X2, Y2, Z2, zdir='z', s=10, c='b', depthshade=True)
    ax.set_xlabel("X",labelpad=2)
    ax.set_ylabel("Y",labelpad=2)
    ax.set_zlabel("Z")
    print(ax.spines)
    ax.spines['top'].set_position(('data', 0))
    plt.show()
```

```python
def ex1_6(u,v,n_samples=200):
    t = np.expand_dims(np.linspace(-5,5,num=n_samples),axis = 0)
    u = np.expand_dims(u,axis=0)
    matrix =  np.transpose(t)*u + v
    matrix = np.append(matrix,[u[0],v],axis = 0)
    return matrix
x = ex1_6(np.array([0,1,0]),np.array([2,3,4]),n_samples=50)
print(x.shape)
illustration_3d_func(x)
```

```python
def ex1_7(u,v,n_samples=200):
    t = np.expand_dims(np.linspace(-25,25,num=n_samples),axis = 0)
    u = np.expand_dims(u,axis=0)
    matrix = np.transpose(t)*u + (1-np.transpose(t))*v
    matrix = np.append(matrix,[u[0],v],axis = 0)
    return matrix
x = ex1_7(np.array([1,2,3]),np.array([2,3,4]),n_samples=50)
illustration_3d_func(x)
```

```python
def ex1_9(u,n_samples=200):
    t = np.expand_dims(np.linspace(0,1,num=n_samples),axis = 0)
    u = np.expand_dims(u,axis=0)
    matrix = np.transpose(t)*u
    matrix = np.append(matrix,u,axis = 0)
    return matrix
x = ex1_8(np.array([1,2,3]),np.array([2,3,4]),n_samples=50)
illustration_3d_func(x)
```

```python
def ex1_10(u,v,n_samples=200):
    t = np.expand_dims(np.linspace(0,1,num=n_samples),axis = 0)
    u = np.expand_dims(u,axis=0)
    matrix = np.transpose(t)*u + (1-np.transpose(t))*v
    matrix = np.append(matrix,[u[0],v],axis = 0)
    return matrix
x = ex1_10(np.array([1,2,3]),np.array([2,3,4]),n_samples=50)
illustration_3d_func(x)
```

```python
def ex1_11(u,v,n_samples=50):
    t = np.expand_dims(np.linspace(0,1,num=n_samples),axis = 0)
    print("u,t,v",u.shape,t.shape,v.shape)
    matrix = np.stack([t[0,i]*u + (1-t[0,i])*v for i in range(t.shape[1])],axis = 0)
    matrix = np.reshape(matrix,newshape=(-1,3))
    print(matrix.shape)
    return matrix
y =ex1_11(np.array([3,5,10]),ex1_11(np.array([0,1,0]),ex1_11(np.array([1,0,0]),np.array([0,0,1]))),n_samples=50)
illustration_3d_func(y)
```
